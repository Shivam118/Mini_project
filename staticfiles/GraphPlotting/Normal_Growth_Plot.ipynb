```python
%matplotlib inline

import matplotlib.pyplot as plt
# import matplotlib.animation as ani
import numpy as np
import pandas as pd

#Country_Name as Input
country_name = input("Country: ")

#File Path
file_path = "DeathAndGrowthRate.csv" + '\\' + country_name.upper() + '.csv'
data = pd.read_csv(file_path)

#assigning Particular columns to its variable
year = data['Year']
growth_rate = data['Growth Rate']

#converting it to array
year = year.to_numpy()
growth_rate = growth_rate.to_numpy()

#Plotting Graph
plt.plot(year,growth_rate,label="Growth Rate",linewidth="1",color='g')
plt_title = 'Growth in ' + country_name
plt.legend()  #units
plt.title(plt_title)  #title of Graph
plt.xlabel('Year')   #X-Axis Label
plt.ylabel("RATE (in %age)")   #Y-Axis Label

plt.savefig(r'../App/Static/Files/plot.png',dpi=100)

plt.show()
```
